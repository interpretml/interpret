```python
from interpret.glassbox._ebm._excel import UNTESTED_to_excel_exportable
```

```python
import pickle
```

```python
with open("titanic.pkl", "rb") as file:
    ebm = pickle.load(file)
```

```python
ebm
```

```python
model_description = "Titanic dataset – EBM Model"
variables_description = {
    'Pclass' : "whether the passenger was traveling in first, second, or third class",
    'Sex': "male or female",
    'Age': "age of passenger",
    'SibSp': "number of siblings aboard the ship",
    'Parch': "number of parents or children aboard the ship",
    'Fare': "price paid for fare",
}
default_evaluation = {
    'Pclass': 3,
    'Sex': "male",
    'Age': 6.0,
    'SibSp': 0,
    'Parch': 1,
    'Fare': 12.475,
}
```

```python
workbook = UNTESTED_to_excel_exportable(
    ebm=ebm,
    file="to_excel.xlsx",
    model_description=model_description,
    variables_description=variables_description,
    default_evaluation=default_evaluation,
)
```

```python
workbook.close()
```
